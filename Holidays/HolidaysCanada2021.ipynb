### Create Canada holiday table by province Vertically

```python
holidays_ca = pd.DataFrame(columns=['holidays', 'holiday_names', 'province/state', 'country'], index=None)
for prov in (['AB', 'BC', 'MB', 'NB', 'NL', 'NS', 'NT', 'NU', 'ON', 'PE', 'QC', 'SK', 'YU' ]):
    canada_holidays = holidays.CA(years=2021, prov=prov).items()
    canada_holidays = pd.DataFrame(data=canada_holidays, columns=['holidays', 'holiday_names'])
    canada_holidays['province/state'] = prov
    canada_holidays['country'] = 'Canada'
    holidays_ca = pd.concat([holidays_ca, canada_holidays]).reset_index(drop=True)
holidays_ca
```

```python
holidays_ca.to_csv('HolidaysCA2021V.csv')
```

### Create Canada holiday table by province Horizontally

```python
holidays_ca = pd.DataFrame(columns=['holidaysDate'])
for prov in (['AB', 'BC', 'MB', 'NB', 'NL', 'NS', 'NT', 'NU', 'ON', 'PE', 'QC', 'SK', 'YU' ]):
    canada_holidays = holidays.CA(years=2021, prov=prov).items()
    canada_holidays = pd.DataFrame(data=canada_holidays, columns=['holidaysDate', 'province'+prov])
    holidays_ca = pd.merge(holidays_ca, canada_holidays, how='outer', on='holidaysDate').set_index('holidaysDate')
holidays_ca.sort_values('holidaysDate', inplace=True)
holidays_ca
```

```python
holidays_ca.to_csv('HolidaysCA2021H.csv')
```
